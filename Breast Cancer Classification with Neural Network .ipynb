```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets

from sklearn.model_selection import train_test_split
```

```python
# loading the data 
breast_cancerdataset = sklearn.datasets.load_breast_cancer()
```

```python
#display 
print(breast_cancerdataset)
```

```python
# loads the data
dataframe = pd.DataFrame(breast_cancerdataset.data, columns = breast_cancerdataset.feature_names)
# displays the top 5 records
dataframe.head()
```

```python
# adds the 'target' column to the dataframe
dataframe['label'] = breast_cancerdataset.target
#print the last 5 records
dataframe.tail()
```

```python
# displays the shape of the dataset
dataframe.shape
```

```python
# displays information
dataframe.info()
```

```python
# displays missing values
dataframe.isnull().sum()
```

```python
# displays the statistics of the dataset
dataframe.describe()
```

```python
# checks the distribution of Target Varibale

'''
Benign=1
Malignant=0
'''
dataframe['label'].value_counts()
```

```python
#displays the mean 
dataframe.groupby('label').mean()
```

```python

# Remove the 'label' column from the dataframe and assign the result to 'X'.
x = dataframe.drop(columns='label', axis=1)

# Select the 'label' column from the dataframe and assign it to 'Y'.
y = dataframe['label']

print('X:',x)
print('Y:',y)
```

```python
#splits into train & test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
print(x.shape, x_train.shape, x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Create a StandardScaler object, which will standardize the features by removing the mean and scaling to unit variance.
scaler = StandardScaler()
# Standardize the training data (x_train) using the scaler, and store the result in x_train_std.
x_train_std = scaler.fit_transform(x_train)
# Standardize the test data (x_test) using the same scaler that was fit to the training data, and store the result in x_test_std.
x_test_std = scaler.transform(x_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)

from tensorflow import keras
```

```python
# Setting up the layers of a Neural Network using Keras.
# Create a Sequential model.
model = keras.Sequential([
    # Flatten layer to transform the input data shape (30,) to a flat vector.
    keras.layers.Flatten(input_shape=(30,)),
    # Dense hidden layer with 20 neurons and ReLU activation function.
    keras.layers.Dense(20, activation='relu'),
    # Output layer with 2 neurons and sigmoid activation for binary classification.
    keras.layers.Dense(2, activation='sigmoid')
])
# Compiling the Neural Network model using the Adam optimizer,
# sparse categorical cross-entropy loss, and accuracy as a metric.
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the network
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=34)
```

```python
# Plotting the model's training and validation accuracy over epochs.
plt.plot(history.history['accuracy'], color='blue', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], color='green', label='Validation Accuracy')
plt.title('Model Accuracy Over Epochs')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
```

```python
# Plotting the model's training and validation loss over epochs.
plt.plot(history.history['loss'], color='red', label='Training Loss')
plt.plot(history.history['val_loss'], color='purple', label='Validation Loss')
plt.title('Model Loss Over Epochs')  # Added title.
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')  # Changed legend location.
```

```python
# Evaluating the model's accuracy on the test dataset and printing the result.
loss, accuracy = model.evaluate(x_test_std, y_test)
print("Test Accuracy: {}".format(accuracy),x_test_std.shape,x_test_std[0])
print('\nAccuracy:',accuracy)
```

```python
# Making predictions on the test data using the trained model and printing the results.
ypred = model.predict(x_test_std)
print(f"Predictions: Shape - {ypred.shape}, First - {ypred[0]}, All - {ypred}")

# Finding the index of the maximum value in the list 'mylist' using np.argmax().
mylist = [0.25, 0.56]
maxvalueindex = np.argmax(mylist)
print(mylist,maxvalueindex)
# Converting prediction probabilities to class labels using np.argmax().
ypredlabels = [np.argmax(i) for i in ypred]
print(ypredlabels)
```

```python
# Input data for prediction
inputData = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)
# Convert the input data to a numpy array
input_asnumpyarray = np.asarray(inputData)
# Reshape the numpy array as we are predicting for one data point
dataReshaped = input_asnumpyarray.reshape(1, -1)
# Standardize the input data
data_std = scaler.transform(dataReshaped)
# Make predictions using the trained model
prediction = model.predict(data_std)

print(prediction)

# Convert prediction probabilities to class label
predictlabel = [np.argmax(prediction)]
print(predictlabel)
```
